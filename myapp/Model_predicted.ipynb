```python
smote = SMOTE(sampling_strategy='minority')
```

```python
X = np.array(df['vector'].tolist())
```

```python
X_sm, y_sm = smote.fit_resample(X, y)
```

```python
X_model =  pd.DataFrame(X_sm)
```

```python
X_model = X_model.abs()
```

```python
y_model = pd.DataFrame(y_sm)
```

```python
type(y_model)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_model, y_model, test_size=0.3, random_state=42)

# Train a Naive Bayes classifier on the vectorized data
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)

y_pred = nb_classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
import pickle

# Save the model as a pickle file
with open('nb_classifier.pkl', 'wb') as file:
    pickle.dump(nb_classifier, file)
```

```python
import pickle

# Load the saved model from the pickle file
with open('nb_classifier.pkl', 'rb') as file:
    nb_classifier = pickle.load(file)

# Use the trained model for prediction
y_pred = nb_classifier.predict(X_test)
```
